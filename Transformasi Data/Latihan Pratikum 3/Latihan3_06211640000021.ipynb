```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("pcademo.csv")
data.head()
```

```python
#cara mengambil data tertentu saja/memotong data
x=data.iloc[0:51,1:20]
x.tail()
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') #default norm l2, other are l1 and max
xnorm=pd.DataFrame(xnorm)
xnorm.tail()
```

```python
#minmax
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-5,1))
scaler.fit(x)
x_scaled=scaler.transform(x)
x_scaled=pd.DataFrame(x_scaled)
x_scaled.tail()
```

```python
#3igmoid
xsigmoidal=(1-2.718281818**(-xnorm))/(1+2.718281818**(-xnorm))
xsigmoidal=pd.DataFrame(xsigmoidal)
xsigmoidal.tail()
```

```python
#desimal scalling
xdecimal=x/(10**5)
xdecimal=pd.DataFrame(xdecimal)
xdecimal.tail()
```

```python
#softmax (fungsi dibuat sendiri)
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis=0)
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
xsoft.tail()
```

```python
#feature selection (mencari skor)
from sklearn.feature_selection import SelectKBest, chi2, f_regression
Xx=x.iloc[0:51,1:20]
y=x.iloc[0:51,1]
selector=SelectKBest(f_regression,k="all").fit(Xx,y)
X_new= selector.transform(Xx)
scores= selector.scores_
scores
```
