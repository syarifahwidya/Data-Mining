## Data Integration

```python
import pandas as pd
```

```python
#Memanggil file yang sudah terdapat pada komputer
data1=pd.read_csv("Midwest1.csv", sep=";")
data2=pd.read_csv("Midwest2.csv",sep=";")

#mencetak data
print(data1)
print(data2)
```

```python
#peggabungan kolom menggunakan fungsi Merge
#menggabungkan data dengan adanya kata kunci pada kedua data yang ingin digabungkan

data_complate=pd.merge(data1, data2, on='county')
data_complate.head()
```

```python
#peggabungan kolom menggunakan fungsi concat
#menggabungkan data secara langung tanpa perlu adanya kata kunci pada data
data_complate1= pd.concat([data1, data2], axis=1)
data_complate1.head()
```

```python
#Memanggil file yang sudah terdapat pada komputer
data3=pd.read_csv("Midwest3.csv", sep=";")
data4=pd.read_csv("Midwest4.csv", sep=";")
data5=pd.read_csv("Midwest5.csv",sep=";")

#Mencetak data
print(data3)
print(data4)
print(data5)
```

```python
#Peggabungan baris dengan menggunakan fungsi concat
#menggabungkan data dengan setiap nama kolom harus sama

frames=[data3,data4,data5]
data_complate2=pd.concat(frames)
data_complate2
```

```python
#Peggabungan baris dengan menggunakan fungsi append
#menggabungkan data dengan setiap nama kolom harus sama
data_complate3=data3.append([data4,data5], sort=False)
data_complate3
```
