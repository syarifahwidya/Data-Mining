## Missing Value

```python
#Pengecekan missing value
np.sum(data.isnull())
```

```python
#melakukan imputasi dengan menggunakan nilai mean
#data berdistribusi normal, karena nilai mean dan median hampir sama
data['MinTemp']=data['MinTemp'].fillna (data['MinTemp'].mean())
data['MaxTemp']=data['MaxTemp'].fillna (data['MaxTemp'].mean())
data['Evaporation']=data['Evaporation'].fillna (data['Evaporation'].mean())
data['Sunshine']=data['Sunshine'].fillna (data['Sunshine'].mean())
data['WindGustSpeed']=data['WindGustSpeed'].fillna (data['WindGustSpeed'].mean())
data['WindSpeed9am']=data['WindSpeed9am'].fillna (data['WindSpeed9am'].mean())
data['WindSpeed3pm']=data['WindSpeed3pm'].fillna (data['WindSpeed3pm'].mean())
data['Humidity9am']=data['Humidity9am'].fillna (data['Humidity9am'].mean())
data['Humidity3pm']=data['Humidity3pm'].fillna (data['Humidity3pm'].mean())
data['Pressure9am']=data['Pressure9am'].fillna (data['Pressure9am'].mean())
data['Pressure3pm']=data['Pressure3pm'].fillna (data['Pressure3pm'].mean())
data['Cloud9am']=data['Cloud9am'].fillna (data['Cloud9am'].mean())
data['Cloud3pm']=data['Cloud3pm'].fillna (data['Cloud3pm'].mean())
data['Temp9am']=data['Temp9am'].fillna (data['Temp9am'].mean())
data['Temp3pm']=data['Temp3pm'].fillna (data['Temp3pm'].mean())

##melakukan imputasi dengan menggunakan nilai median
#data tidak berdistribusi normal
data['Rainfall']=data['Rainfall'].fillna (data['Rainfall'].median())
data['RISK_MM']=data['RISK_MM'].fillna (data['RISK_MM'].median())
```

```python
#Imputasi dengan menggunakan modus untuk data kategorik

from scipy.stats import mode
data['WindGustDir']=data['WindGustDir'].fillna(data['WindGustDir']).dropna().mode()[0]
data['WindDir9am']=data['WindDir9am'].fillna(data['WindDir9am']).dropna().mode()[0]
data['WindDir3pm']=data['WindDir3pm'].fillna(data['WindDir3pm']).dropna().mode()[0]
data['RainToday']=data['RainToday'].fillna(data['RainToday']).dropna().mode()[0]
data['RainTomorrow']=data['RainTomorrow'].fillna(data['RainTomorrow']).dropna().mode()[0]
```

```python
#Mengecekan ulang missing value
np.sum(data.isnull())
```

## Outlier

```python
from scipy import stats
datanumerik=data.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow'],1)
z = np.abs(stats.zscore(datanumerik._get_numeric_data()))
print(z)

dataoutlier=datanumerik[(z<3).all(axis=1)]
print(dataoutlier.shape)
```

```python
dataoutlier
```

# Transformasi

```python
datatrans=dataoutlier.iloc[0:134754,0:18]
```

```python
#Melakukan transformasi dengan menggunakan metode normalization untuk data numerik
from sklearn.preprocessing import normalize
datnorm=normalize(datatrans,norm='l2')
datnorm=pd.DataFrame(datnorm)
print(datnorm.shape)
datnorm.head()
```

## Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2, f_regression
datanormX=datnorm.iloc[0:134754,0:18]
y=datnorm.iloc[0:134754,1]
selector=SelectKBest(f_regression,k="all").fit(datanormX,y)
X_new= selector.transform(datanormX)
scores= selector.scores_
scores
```
